```python
import requests
import csv
```

```python
response = requests.get('https://www.metaweather.com/api/location/1521894/')
response
```

```python
response.json()
```

```python
weather = response.json()
weather
```

```python
weather['consolidated_weather']
```

```python
for w in weather['consolidated_weather'] :
    print(w['applicable_date'] , '|' , w['created'] , '|' , w['the_temp'] , '|' , w['humidity'] , '|' , w['visibility'] , '|' , w['air_pressure'] , '|' , w['wind_speed'])
```

```python
with open('weather.csv' , 'w' , newline='') as f:
    writer = csv.DictWriter(f , fieldnames=['applicable_date' , 'created' , 'the_temp' , 'humidity' , 'visibility' , 'air_pressure' , 'wind_speed' ])
    writer.writeheader()

    for w in weather['consolidated_weather'] :
        applicable_date_ = w['applicable_date']
        created = w['created']
        the_temp = w['the_temp']
        humidity = w['humidity']
        visibility = w['visibility']
        air_pressure = w['air_pressure']
        wind_speed = w['wind_speed']

        writer.writerow(
            { 'applicable_date' : applicable_date_ ,
            'created' : created ,
            'the_temp' : the_temp , 
            'humidity' : humidity , 
            'visibility' : visibility , 
            'air_pressure' : air_pressure ,
            'wind_speed' : wind_speed
            }
            )

```
